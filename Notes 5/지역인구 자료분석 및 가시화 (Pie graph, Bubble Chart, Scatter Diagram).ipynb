```python
# pie() 함수 사용하기
import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
plt.axis('equal')        # pie graph 의 반지름을 균일하게 (원형이 안 찌그러지도록)
plt.pie(size)
plt.show()
```

```python
# pie 함수에 label 추가하기, color 를 임의로 지정하기.
import matplotlib.pyplot as plt
plt.rc('font', family = 'Apple SD Gothic Neo')
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'AB형', 'O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.axis('equal')
plt.pie(size, labels = label, autopct='%.1f%%', colors=color, explode=(0,0,0.1, 0))
'''
pie 함수의 option들:
size: graph data
labels : label data
autopct : 비율 표시 (ex) 소수점 아래 1번째 자리까지)
colors : color data (https://matplotlib.org/gallery/color/named_colors.html)
explode : 어떤 값을 highlight하기 위해 사용가능한 option. default = (0,0,0,0)
'''
plt.legend()
plt.show()
```

```python
# 성별 연령별 인구 데이터 출력 (pie chart)
import csv
f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 입력하세요 : ')
for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)
plt.rc('font', family = 'Apple SD Gothic Neo')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['남','여'], autopct = '%.1f%%', colors = color, startangle=90)
#startangle parameter : (default=0) pie chart에서 데이터 아이템 slice시작을 어디서 할지 (x축기준+angle)
# ex) startangle=0 : 첫번째 아이템 start 위치를 x축에서 시작
# ex2) startangle=90 : default 기준에서 반시계방향으로 90도 회전
plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```

```python
# 지역별 연령별 (남성인구-여성인구) 인구 값 막대그래프로 표현
import csv
f = open('gender.csv',encoding = 'cp949')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i])- int(row[i+103]))
        break     #혹시나 중복이 있다면 첫번째 지역에서만 찾는다

import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
plt.figure(dpi=300)
plt.bar(range(101),result)
plt.show()
```

```python
# 제주도의 연령대 별 성별 비율을 꺾은선 그래프로 표현
import csv
f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
```

```python
# 제주도의 연령대 별 성별 비율을 버블차트로 표현
import csv
f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')     # 격자무늬 추가
plt.scatter(m,f)
plt.show()
```

```python
# 제주도의 연령대 별 성별 비율을 산점도로 표현 (원의 크기는 인구 수의 제곱근)
import csv
import math
file = open('gender.csv', encoding = 'cp949')  # 이후에 사용되는 변수명과 다르게 임의로 파일의 변수명을 바꾸었습니다.
data = csv.reader(file)
m = []
f = []
size = []       # 연령별 인구수를 나타내는 점의 크기
name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103])))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')           # 격자무늬 추가
plt.rc('font', family = 'Apple SD Gothic Neo')
plt.figure(figsize=(10,5), dpi=300)
plt.scatter(m,f, s = size, c = range(101), alpha = 0.5, cmap = 'jet')
# data =m,f, dot_size data = size, color_data = 0~100, transparency = 0.5, cmap = 'jet'
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')   # 추세선 표시 (연령별 남성 인구 최대 값 기준: 두 인구 중 더 큰 남성인구 기준)
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
file.close()
```
