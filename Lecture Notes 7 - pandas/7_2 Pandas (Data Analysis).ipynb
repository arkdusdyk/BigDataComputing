```python
# Exercise 3 : using pandas to visualize population structure
# Numpy 로 구현 해본것 pandas로 구현해보기
import pandas as pd
df = pd.read_csv('age.csv',encoding = 'cp949', index_col=0) # index_col = 0 : DataFrame 의 label로 사용할 row 의 index지정
df.head()
```

```python
# 전체 데이터 비율로 변환, 총 인구수, 연령 구간 인구수 열 삭제
import pandas as pd
df = pd.read_csv('age.csv',encoding = 'cp949', index_col=0) # index_col = 0 : DataFrame 의 label로 사용할 row 의 index지정
df = df.div(df['총인구수'], axis=0)   # 전체 데이터를 총 인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']
df.head()
```

```python
# 궁금한 지역 이름 입력, 해당 지역 인구 구조 저장
name = input('원하는 지역 이름 입력 : ')
a = df.index.str.contains(name)
df2 = df[a]
df2
```

```python
# Visualize data using Graph
import matplotlib.pyplot as plt
plt.rc('font', family = 'Apple SD Gothic Neo')
df2.T.plot()   #Transpose df2 (연령이 x축에 오도록)
plt.show()
```

```python
# 궁금한 지역의 인구구조 비율과 차이 계산하기
import numpy as np
x = df.sub(df2.iloc[0], axis=1)   #궁금한 지역의 인구 비율에서 현재 지역 인구 비율 뺀다.(행단위 연산)
y = np.power(x,2)     # numpy 의 power()를 통해 이 뺀 값들의 제곱값들을 찾아서
z = y.sum(axis=1)     # 모두 더한다
# now find 2nd minimum of z (1st minimum : query row)
```

```python
# Full Exercise3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family = 'Apple SD Gothic Neo')
df = pd.read_csv('age.csv', encoding = 'cp949', index_col=0)
df = df.div(df['총인구수'], axis=0)
del df['총인구수'], df['연령구간인구수']

name = input('원하는 지역의 이름을 입력하세요: ')
a = df.index.str.contains(name)
df2 = df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
#Numpy로 구현했을때와 차이를 구별하기!!
```
