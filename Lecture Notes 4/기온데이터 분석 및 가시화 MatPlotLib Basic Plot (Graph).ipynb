```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
result = []
for row in data :
    if row[-1] != '':                    #     ,
        if row[0].split('-')[1]=='08':   # 8  ,
            result.append(float(row[-1]))    # result     
plt.plot(result, 'hotpink')   # result    hotpink 
plt.show()
```
