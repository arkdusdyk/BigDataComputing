```python
# 동네 입력 받아서 인구 구조 (연령별 인구수 데이터 시각화)
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)
dong = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ")
result = []
for row in data:
    if dong in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
import matplotlib.pyplot as plt
plt.style.use('ggplot')    # 격자 무늬 스타일
plt.plot(result)
plt.show()
f.close()
```

```python
# 막대그래프
# bar(막대를 표시할 위치, 막대의 높이)
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]: # row[3:] : 0세부터 100세 연령별 인구수
            result.append(int(i))
import matplotlib.pyplot as plt
plt.bar(range(101),result)    # x축: 연령(0~100), y축 : 연령별 인구수
plt.show()
```

```python
# 막대그래프
# bar(막대를 표시할 위치, 막대의 높이) - vertical(default)
import csv
f = open('age.csv', encoding = 'cp949')
data = csv.reader(f)
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[3:]: # row[3:] : 0세부터 100세 연령별 인구수
            result.append(int(i))
import matplotlib.pyplot as plt
plt.barh(range(101),result)    # barh()함수 : horizontal (y축, x축)
plt.show()
```

```python
# 항아리 모양 그래프 그리기 (성별)
# 남성 데이터 : 3~103번
# 여성 데이터 : -1~ -101번
import csv
file = open('gender.csv', encoding = 'cp949')
data = csv.reader(file)
m = []
f = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()    #f는 현재 -1, -2, -3 순..
#그냥 m은 row[3:104], f는 row[106:] 로 둬서 reverse안해도 됨..
import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
file.close()
#겹쳐서 나오는 단점-> 해결? 남성 데이터를 음수로
```

```python
# 항아리 모양 그래프 그리기 (성별)
# 남성 데이터 : 3~103번
# 여성 데이터 : -1~ -101번
import csv
file = open('gender.csv', encoding = 'cp949')
data = csv.reader(file)
m = []
f = []
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(-(int(row[i+3]))) ####마이너스 부호를 통해 남성데이터를 음수로 바꾸기..
            f.append(int(row[-(i+1)]))
f.reverse()


import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family = 'Apple SD Gothic Neo')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
file.close()
```
