```python
import matplotlib.pyplot as plt    # pyplot 모듈 불러오기
plt.title('plotting')              # 그래프에 제목 달기
plt.plot([1,2,3,4],[12,43,25,15])  # plt.plot([x 좌표들], [y 좌표들]) -> 실제로 찍힐 때는 실수값으로 찍힌다..
plt.show                           # plt.show() : display all open figures
```

```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40],label = 'asc') # 만약 plt.plot에 list가 한개면, 자동으로 그 list가 y값, list의 index가 x값이 됨
plt.plot([40,30,20,10],label = 'desc')
plt.legend()                          #legend (범례 : 선 labeling)
plt.show()
```

**만약 plt.plot에 list가 한개면, 자동으로 그 list가 y값, list의 index가 x값이 됨

```python
#그래프 색상 바꾸기
import matplotlib.pyplot as plt
plt.title('color')
# 색깔을 바꾸는 두 가지 방법
plt.plot([10,20,30,40], color = 'skyblue', label = 'skyblue')
plt.plot([40,30,20,10], 'pink', label = 'pink')
plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label = 'circle')    # 'r.' = (color=red, marker=dot)
plt.plot([40,30,20,10],'g^',label = 'triangle up') # 'g^' = (color=green, marker=triangle)
plt.plot([25,25,25,25], 'b*', label = 'star')
plt.legend()
plt.show()
```
