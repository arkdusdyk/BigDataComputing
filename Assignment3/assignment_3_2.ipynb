```python
# 과제 3-1에 "제주특별자치도"를 입력값으로 받아서 항아리 모양 그래프 출력
import csv
import matplotlib.pyplot as plt
file = open('gender.csv', encoding = 'cp949')
data = csv.reader(file)
region = input("찾고 싶은 지역의 이름을 알려주세요 : ")
m = []
f = []
for row in data:
    if region in row[0]:
        for i in range(0,101):
            m.append(-(int(row[i+3])))
            f.append(int(row[-(i+1)]))
        break
f.reverse()

plt.rc('font', family = 'Apple SD Gothic Neo')
plt.figure(figsize=(10,5), dpi=300)          #figure size, dpi 조절
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')    # 격자 무늬 스타일
plt.title(region+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
file.close()
```

```python
# 제주도의 연령대 별 성별 비율을 산점도로 표현 (원의 크기는 인구 수의 제곱근)
import csv
import math
file = open('gender.csv', encoding = 'cp949')  # 이후에 사용되는 변수명과 다르게 임의로 파일의 변수명을 바꾸었습니다.
data = csv.reader(file)
m = []
f = []
size = []       # 연령별 인구수를 나타내는 점의 크기
name = input('궁금한 동네를 입력해주세요 : ')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103])))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')           # 격자무늬 추가
plt.rc('font', family = 'Apple SD Gothic Neo')
plt.figure(figsize=(10,5), dpi=300)
plt.scatter(m,f, s = size, c = range(101), alpha = 0.5, cmap = 'jet')
# data =m,f, dot_size data = size, color_data = 0~100, transparency = 0.5, cmap = 'jet'
plt.colorbar()
plt.title(name+' 지역의 성별 인구 그래프')
plt.plot(range(max(m)), range(max(m)), 'g')   # 추세선 표시 (연령별 남성 인구 최대 값 기준: 두 인구 중 더 큰 남성인구 기준)
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
file.close()
```
